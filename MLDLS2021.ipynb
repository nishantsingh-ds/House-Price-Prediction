**Import Libraries**

```python
#importing all necessary libraries required.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

**Import Dataset**

```python
dataset=pd.read_csv('train2.csv')

#print shape of dataset with rows and columns
print(dataset.shape)
```

```python
dataset
```

```python
Y = dataset.iloc[:,-1].values
X = dataset.iloc[:,19].values

df = pd.DataFrame(X,Y)
print(df)
```

```python
plt.xlabel("Sale Price")
plt.ylabel("Year Built")
plt.title("Year Built Vs Sale Price")
plt.scatter(X,Y, color='red', marker='*')
```

**Split the data**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3)
print(X_train.shape, X_test.shape)
```

```python
X_train = X_train.values.reshape(-1, 1)
Y_train= Y_train.values.reshape(-1, 1)
X_test = X_test.values.reshape(-1, 1)
```

**Fitting Simple Linear Regression**

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train, Y_train)
```

**Predicting the test Results**

```python
Y_pred = model.predict(X_test)
```

**Calculating Accuracy**

```python
accuracy=model.score(X_test,Y_test)
print(accuracy)
```

**Plot the Regressor Curve**

```python
plt.scatter(X_train,Y_train,color='red',marker='+')
plt.xlabel('Sale Price')
plt.ylabel('Year Built')
plt.title('Sale Price Vs Year Built (Training Set)')
plt.plot(X_train, model.predict(X_train), color='blue')
plt.show()

plt.scatter(X_test,Y_test,color='red',marker='+')
plt.xlabel('Sale Price')
plt.ylabel('Year Built')
plt.title('Sale Price Vs Year Built (Training Set)')
plt.plot(X_test, model.predict(X_test), color='blue')
plt.show()
```

**Prediction on External Data**

```python
test_data = np.array([[2021]])
test_pred = model.predict(test_data)
print(test_pred)
```
